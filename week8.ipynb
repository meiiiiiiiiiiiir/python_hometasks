```python
import pandas as pd
#1.Адамның аты мен жасынан кесте құр.
df=pd.DataFrame({"Aты":["Айжан","Мұрат","Санжар"], "Жасы":[20,22,19]})
print(df)
#2.Кестеден тек “Аты” бағанын шығарып көрсет.
print(df["Aты"])
#3.Кестеге “Баға” бағанын қосып, мәндерін енгіз
df["marks"]=[85,90,75]
print(df)
#4.“Баға” бағанының орташа мәнін есепте.
print(df["marks"].mean())
#5.Кімнің бағасы ең жоғары екенін анықта.
bestmarks=df[df["marks"]==df["marks"].max()]
print(bestmarks)
#6.Бағасы 80-нен жоғары студенттерді көрсет.
over80=df[df["marks"]>80]
print(over80)
#7.Кестені “Баға” бойынша кему ретімен сұрыпта
sorteddf=df.sort_values(by="marks")
print(sorteddf)
#8.Кестенің индекстерін 0, 1, 2 деп қайта орнат.
print(df.reset_index(drop=True))
#9.“Жасы” бағанын өшіріп таста.
print(df.drop(columns=["Жасы"]))
#10.“Баға” бағанын “Ұпай” деп өзгерт.
df=df.rename(columns={"marks":"assessments"})
print(df)
#11.Тек “Аты” мен “Баға” бағандарын көрсет.
print(df[["Aты","assessments"]])
#12.Кестенің алғашқы екі қатарын (head(2)) көрсет.
print(df.head(2))
#13.Кестенің соңғы екі қатарын (tail(2)) көрсет.
print(df.tail(2))
#14.Кестедегі сандар бойынша статистиканы (describe()) көрсет.
print(df.describe())
#15.Бағасы орташа мәннен жоғары студенттерді тап.
mean=df["assessments"].mean()
print(df[df["assessments"]>mean])
#16.“Баға + 5” деп жаңа баған қос.
df["assessments+5"]=df["assessments"]+5
print(df)
```

```python
#17.2 түрлі топтың студенттерін көрсететін екі кесте құр:
import pandas as pd
a=pd.DataFrame({"Name":["Aizhan","Murat"],"marks":[85,90]})
b=pd.DataFrame({"Name":["Sanzhar","Aset"],"marks":[80,100]})
print(a)
print(b)
#18.Топ A мен Топ B кестелерін бір үлкен кестеге біріктір (pd.concat()).
print(pd.concat([a,b]))
#19.Бір кестеде “Аты”, екіншісінде “Баға” болса — оларды баған бойынша біріктір.
print(pd.concat([a["Name"],b["marks"]],axis=1))
```

```python
#20.Егер кейбір жолдар қайталанса — оларды өшір (drop_duplicates()).
import pandas as pd
df=pd.DataFrame({"Name":["Aizhan","Sanzhar","Sanzhar"],"marks":[None,90,90]})
print(df.drop_duplicates())
#21.Кестеде бос (NaN) мәндер бар ма, тексер.
print(df.isna())
```

```python
#22.NaN мәндерінің орнына орташа мәнді қой (fillna()).
import pandas as pd
df=pd.DataFrame({"Name":["Aizhan","Sanzhar","Sanzhar"],"marks":[None,85,90]})
df["marks"]=df["marks"].fillna(df["marks"].mean())
print(df)
```

```python
#23.Баға 70 пен 90 аралығында болған студенттерді тап.
import pandas as pd
df=pd.DataFrame({"Name":["Aizhan","Sanzhar","Sanzhar"],"marks":[95,85,90]})
print(df[(df["marks"]>=70) & (df["marks"]<=90)])
#24.Кестеден екінші жолды (iloc[1]) көрсет.
print(df.iloc[1])
```

```python
#25.“Жасы” бағанын бүтін (int) типінен нақты (float) типіне ауыстыр.
import pandas as pd
df=pd.DataFrame({"Aты":["Айжан","Мұрат","Санжар"], "Жасы":[20,22,19]})
df["Жасы"]=df["Жасы"].astype(float)
print(df.dtypes)
```

```python
#26.Барлық бағаларға 10% қос (apply(lambda x: x * 1.1)).
import pandas as pd
df=pd.DataFrame({"Name":["Aizhan","Sanzhar","Sanzhar"],"marks":[75,85,90]})
df["marks"]=df["marks"].apply(lambda x:x*1.1)
print(df)
```

```python
#27.Орташа ұпайдан жоғары болса “Жақсы”, әйтпесе “Орташа” деп баған қос.
import pandas as pd
df=pd.DataFrame({"Name":["Aizhan","Sanzhar","Sanzhar"],"marks":[75,85,90]})
df["status"] = df["marks"].apply(lambda x: "Good" if x >df["marks"].mean() else "OK")
print(df)
```
