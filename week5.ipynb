```python
#Proc19.
def RingS(R1, R2):
    S1=math.pi*R1**2
    S2=math.pi*R2**2
    return abs(S1-S2)

print(RingS(7,5))
print(RingS(6,1))
print(RingS(9,8))
```

```python
#Proc20.
def TriangleP(a,h):
    b=math.sqrt((a/2)**2+h**2)
    return a+b+b
print(TriangleP(10,12))
print(TriangleP(5,9))
print(TriangleP(9,4))
```

```python
#Proc21.
def SumRange(A, B):
    s=0
    if A>B:
        return 0
    else:
        for i in range(A,B+1):
            s=s+i
        return s



print(SumRange(5,9))
print(SumRange(9,1))
print(SumRange(2,71))
```

```python
#Proc22.
def Calc(A, B, op):
    if op=="*":
        return A*B
    elif op=="/":
        return A/B
    elif op=="-":
        return A-B
    elif op=="+":
        return A+B
    else:
        return "wrong operation"


print(Calc(5,9,"*"))
print(Calc(5,9,"-"))
print(Calc(24,6,"/"))
```

```python
#Proc23.
def  Quarter(x, y):
    if x==0 or y==0:
        return "on the axis"
    elif x>0 and y>0:
        return "1st quarter"
    elif  x<0 and y>0:
        return "2nd quarter"
    elif x<0 and y<0:
        return "3rd quarter"
    else:
        return "4th quarter"


print(Quarter(2,3))
print(Quarter(0,-3))
print(Quarter(1,-6))
```

```python
#Proc30.
def digitn(k,n):
    if len(str(k))<n:
        return -1
    else:
        return str(k)[-n]

print(digitn(12345678,4))
print(digitn(987654321,5))
print(digitn(9876523456789,7))
print(digitn(987,9))
```

```python
#3.
def sum_numbers(lst):
    s=0
    for i in range(0,len(lst)):
        if type(lst[i])==int or type(lst[i])==float:
            s=s+lst[i]
    return s
print(sum_numbers([12,145,"lwjbj]",112.23,"erouvi"]))
```
