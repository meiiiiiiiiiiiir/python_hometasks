```python
def wave(n):
    if n == 0:          # Базовый случай
        print(0)
        return
    else:
        print(n)
        wave(n - 1)
        print(n)
wave(8)
```

```python
def normalize_string(s):
    start = 0
    while start < len(s) and s[start] == " ":
        start += 1
    end = len(s) - 1
    while end >= 0 and s[end] == " ":
        end -= 1
    if start > end:
        return ""
    def to_lower(ch):
        if "A" <= ch <= "Z":
            return chr(ord(ch) + 32)
        return ch
    result = ""
    prev_space = False
    for i in range(start, end + 1):
        ch = s[i]
        if ch == " ":
            if not prev_space:
                result += " "
            prev_space = True
        else:
            result += to_lower(ch)
            prev_space = False
    return result
print(normalize_string("vklsobvdp'oSW:L?JH;ovuahwpsiudhcvaidjbf"))
```

```python
def weighted_average(values, weights):
    total_weight = sum(weights)
    if total_weight == 0:
        return 0

    total = 0
    for v, w in zip(values, weights):
        total += v * w

    return total / total_weight
print(weighted_average([10,20,30],[1,2,5]))
```
