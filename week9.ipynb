```python
                    
#1.without
def count_successful_days(steps_per_day,daily_goal):
    successful_days=0
    for steps in steps_per_day:
        if steps>=daily_goal:
            successful_days+=1
    return successful_days

print(count_successful_days([8000,11000,13000,9500,7500,12000],10000))
```

```python
#1.with
def count_successful_days_builtin(steps_per_day,daily_goal):
    return sum(1 for steps in steps_per_day if steps>=daily_goal)
print(count_successful_days_builtin([8000,11000,13000,9500,7500,12000],10000))
```

```python
#2.without
def filter_files_by_extension(filenames,extension):
    filtered_files=[]
    ext_len=len(extension)
    for filename in filenames:
        if len(filename)>ext_len and filename[-ext_len:]==extension:
            filtered_files.append(filename)
    return filtered_files

print(filter_files_by_extension(["hgfxdrfxhvj.png","pohbhjopop.txt","qwerty.txt","qwertyu.py"],".txt"))
```

```python
#2.with
def filter_files_by_extension_builtin(filenames, extension):
    return [file for file in filenames if file.endswith(extension)]
print(filter_files_by_extension(["hgfxdrfxhvj.png","pohbhjopop.txt","qwerty.txt","qwertyu.py"],".txt"))
```

```python
#3.without
def sum_large_payments(payments,min_amount):
    sum=0
    for payment in payments:
        if payment>min_amount:
            sum+=payment
    return sum

print(sum_large_payments([100.0, 250.5, 80.0, 500], 200))
```

```python
#3.with
def sum_large_payments_builtin(payments, min_amount):
    return sum(payment for payment in payments if payment>=min_amount)
print(sum_large_payments([100.0, 250.5, 80.0, 500], 200))
```

```python
#4.without
def count_status(statuses,success_status):
    successful_statuses=0
    for status in statuses:
        if status==success_status:
            successful_statuses+=1
    return successful_statuses

print(count_status(["done", "failed", "done", "in_progress"], "done"))
```

```python
#4.with
def count_status_builtin(statuses, status):
    return statuses.count(status)
print(count_status(["done", "failed", "done", "in_progress"], "done"))
```

```python
#5.without
def limit_consecutive_repeats(items,max_repeat):
    result=[items[0]]
    consecutive_repeats=1
    for i in range(1,len(items)):
        if items[i]==items[i-1]:
            if consecutive_repeats<max_repeat:
                result.append(items[i])
                consecutive_repeats+=1
        else:
            result.append(items[i])
            consecutive_repeats=1
    return result
print(limit_consecutive_repeats(["a", "a", "a", "b", "b", "a"], 2))
```

```python
 #5.with
def limit_consecutive_repeats(items,max_repeat):
    result=[]
    for item in items:
        if result.count(item)<max_repeat or result[-1]!=item:
            result.append(item)
    return result
print(limit_consecutive_repeats(["a", "a", "a", "b", "b", "a"], 2))
```

```python
#6.without
def suggest_login(taken_logins,base_login):
    if base_login not in taken_logins:
        return base_login
    a=1
    while True:
        login=base_login+str(a)
        if login not in taken_logins:
            return login
        a+=1

print(suggest_login({"user", "admin"}, "newuser"))
```

```python
#8.without
def filter_city_codes_by_country(city_codes,country_prefix):
    result=[]
    prefix_len=len(country_prefix)
    for code in city_codes:
        if len(code)>=prefix_len and code[:prefix_len]==country_prefix:
            result.append(code)
    return result

print(filter_city_codes_by_country({"KZ_ALM", "KZ_AST", "RU_MSK"}, "KZ_"))
```

```python
#8.with
def filter_city_codes_by_country(city_codes, country_prefix):
    return {city for city in city_codes if city.startswith(country_prefix)}
print(filter_city_codes_by_country({"KZ_ALM", "KZ_AST", "RU_MSK"}, "KZ_"))
```

```python
#14.without
def find_first_greater(numbers,x):
    for i in range(len(numbers)):
        if numbers[i]>x:
            return i
    return -1
print(find_first_greater([1, 3, 5, 7], 4)  )
print(find_first_greater([10, 9, 8], 15))
```

```python
#14.with
def find_first_greater(numbers,x):
    for i,number in enumerate(numbers):
        if number>x:
            return i
    return -1
print(find_first_greater([1, 3, 5, 7], 4)  )
print(find_first_greater([10, 9, 8], 15))
```

```python
#15.without
def add_prefix_to_usernames(usernames,prefix):
    result=[]
    prefix_len=len(prefix)
    for username in usernames:
        if len(username)>=prefix_len and username[:prefix_len]==prefix:
            result.append(username)
        else:
            result.append(prefix+username)
    return result
print(add_prefix_to_usernames(["ivan", "vip_olga", "anna"], "vip_"))
```

```python
#17.without
def accept_new_friends(current_friends,new_requests,max_new):
    added_frieds=0
    for i in range(len(new_requests)):
        if new_requests[i] not in current_friends and added_frieds<max_new:
            current_friends.add(new_requests[i])
            added_frieds+=1
    return current_friends
print(accept_new_friends({"ann"}, ["bob", "kate", "max"], 1))
```
