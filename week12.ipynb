```python
#32
def swap_min_max(a):
    min=a[0]
    max=a[0]
    for x in a:
        if x<min:
            min=x
        if x>max:
            max=x
    for i in range(len(a)):
        if a[i]==min:
            a[i]=max
        elif a[i]==max:
            a[i]=min
    print(a)
swap_min_max([1,2,3])
```

```python
#33
def palindrome(n):
    original=n
    reversed=0
    while n>0:
        digit=n%10
        reversed=reversed*10+digit
        n=n//10
    return original==reversed
print(palindrome(123))
print(palindrome(121))
```

```python
#35
def compress(s):
    result=""
    count=1
    for i in range(1,len(s)):
        if s[i]==s[i-1]:
            count+=1
        else:
            result+=s[i-1]+str(count)
            count=1
    result+=s[-1]+str(count)
    print(result)
compress("aaaabbbccc")
compress("pppooooooooooommmmmmm")
```

```python
#47
def bubble(a):
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    print(a)
bubble([2,4,9,3,1,8,6,4,2,6])

```

```python
#58
def most_frequent(a):
    freq={}
    for i in a:
        if i in freq:
            freq[i]=freq[i]+1
        else:
            freq[i]=1
    max=None
    maxval=-1
    for j in freq:
        if freq[j]>maxval:
            maxval=freq[j]
            max=j
    return max
print(most_frequent([1,2,2,3,1,1]))
```

```python
#1
def cold_filter(x,y,flag):
    if flag=="freeze":
        a=x-y
        if a<5:
            return a*10
        else:
            return a*3
    elif flag=="melt":
        if x<y:
            return str(x)+str(y)
        else:
            return str(y)+str(x)
    elif flag=="wind":
        result=[]
        result.append((x+y)/2)
        result.append(abs(x-y))
        return tuple(result)
    else:
        return "wrong flag"

print(cold_filter(10,6,"freeze"))
print(cold_filter(11,2,"melt"))
print(cold_filter(7,12,"wind"))
print(cold_filter(1,6,"asfghk"))
```

```python
def middle_number(a,b,c):
    if a>b and a>c:
        d=b-c
        if d>0:
            return b
        else:
            return c
    elif b>a and b>c:
        d=c-a
        if d>0:
            return c
        else:
            return a
    else:
        if a>b:
            return a
        else:
            return b

print(middle_number(4,8,3))
print(middle_number(4,8,11))
print(middle_number(25,8,12))
```
