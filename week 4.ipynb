```python
#24 Сделать простейший калькулятор в цикле while: +, -, *, /
op=""
while op!="exit":
    op=input("enter operation or exit")
    if op=="exit":
        break
    elif op=="*" or "/" or "+" or "-":
        a=float(input())
        b=float(input())
        if op=="*":
            print(str(a)+op+str(b)+"="+str(a*b))
        elif op=="/":
            print(str(a)+op+str(b)+"="+str(a/b))
        elif op=="-":
            print(str(a)+op+str(b)+"="+str(a-b))
        elif op=="+":
            print(str(a)+op+str(b)+"="+str(a+b))
    else:
        print("wrong operation")
```
