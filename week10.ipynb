```python
#1
def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
print(factorial(5))
print(factorial(30))
```

```python
#2
def sum(n):
    if n==1:
        return 1
    return n+sum(n-1)
print(sum(5))
print(sum(30))
```

```python
#3
def sum_digits(n):
    if n<10:
        return n
    return (n%10)+sum_digits(n//10)
print(sum_digits(598744))
print(sum_digits(308741256985425))
```

```python
#4
def countdown(n):
    if n==1:
        return [1]
    return [n]+countdown(n-1)
print(countdown(5))
print(countdown(30))
```

```python
#5
def reverse_str(s):
    if len(s)<=1:
        return s
    return s[-1]+reverse_str(s[:-1])
print(reverse_str("hello"))
print(reverse_str("jhgfdsa"))
print(reverse_str("ytrewq"))
```

```python
#6
def string_lenght(s):
    if s=="":
        return 0
    return 1+string_lenght(s[1:])
print(string_lenght("hello"))
print(string_lenght("jhgfdsa"))
```

```python
#7
def palindrome(s):
    if len(s)<=1:
        return True
    if s[0]!=s[-1]:
        return False
    return palindrome(s[1:-1])
print(palindrome("hello"))
print(palindrome("aaaaaaaa"))
print(palindrome("aaaaoaaaa"))
print(palindrome("owspieousbgjlvdcxsvof"))
```

```python
#8
def power(a,b):
    if b==0:
        return 1
    return a*power(a,b-1)
print(power(2,3))
print(power(5,7))
print(power(2,4))
```

```python
#9
def list_lenght(l):
    if l==[]:
        return 0
    return 1+list_lenght(l[1:])
print(list_lenght([1,2,3]))
print(list_lenght([1,2,3,1,1,1,1,1,1]))
```

```python
#10
def min_recursive(lst):
    if len(lst)==1:
        return lst[0]
    small=min_recursive(lst[1:])
    if lst[0]<small:
        return lst[0]
    else:
        return small
print(min_recursive([1,2,3]))
print(min_recursive([23,63,32,15,84]))
#23 салыстырады [63,32,15,84]
#63 салыстырады [32,15,84]
#32 салыстырады [15,84]
#15 салыстырады [84]
#84 жалгыз элемент минимум=84
#84 салыстырады 15 минимум=15
#15 салыстырады 32 минимум=15
#15 салыстырады 63 минимум=15
#15 салыстырады 23 минимум=15
```

```python
#11
def remove_all(lst,target):
    if lst==[]:
        return []
    if lst[0]==target:
        return remove_all(lst[1:],target)
    return [lst[0]] +remove_all(lst[1:],target)
print(remove_all([1,2,3,2,3,1,2,4,5,5,4,2,3],3))
```

```python
#12
def fibonachi(n):
    if n==0:
        return 0
    if n==1:
        return 1
    return fibonachi(n-1)+fibonachi(n-2)
print(fibonachi(5))
print(fibonachi(30))
```

```python
def count_digits(n):
    if n<0:
        n=-n
    if 0<=n<=9:
        return 1
    return 1+count_digits(n//10)
print(count_digits(95841625489761526984615))
print(count_digits(874))
```
