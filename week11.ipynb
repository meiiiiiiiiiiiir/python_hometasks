```python
#3
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2 * (self.width + self.height)


r1=Rectangle(10,20)
r2=Rectangle(7,3)

print("Area:",r1.area(),"Perimeter:",r1.perimeter())
print("Area:",r2.area(),"Perimeter:",r2.perimeter())
```

```python
#5
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def deposit(self,amount):
        self.balance += amount
        print(f"replenishment: {amount}")
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"withdrawal: {amount}")
ba1=BankAccount("Max",15000)
ba1.deposit(9000)
ba1.withdraw(10000)
ba1.withdraw(15000)
print("Balance:",ba1.balance)
```

```python
#6
class Car:
    def __init__(self,mark,model,speed):
        self.mark = mark
        self.model = model
        self.speed = speed
    def accelerate(self,x):
        self.speed += x
        print("Speed:", self.speed)
    def brake(self,x):
        if x>self.speed:
            print("Speed:0")
        else:
            self.speed -= x
            print("Speed:", self.speed)
car=Car("Ferrari","F40",220)
car.accelerate(10)
car.accelerate(50)
car.brake(150)
car.brake(999999)
```

```python
#7
class Temperature:
    def __init__(self,temp):
        self.temp = temp
    def to_fahrenheit(self):
        return (self.temp * 9 / 5) + 32
    def to_kelvin(self):
        return (self.temp + 273)
t=Temperature(15)
print("F:",t.to_fahrenheit())
print("K:",t.to_kelvin())
print("C:",t.temp)
```

```python
#8
class Counter:
    def __init__(self,count):
        self.count = count
    def increment(self):
        self.count += 1
        print("count:",self.count)
    def decrement(self):
        self.count -= 1
        print("count:",self.count)
    def reset(self):
        self.count = 0
        print("count:",self.count)
c=Counter(1)
c.increment()
c.decrement()
c.reset()
c.increment()
```

```python
#9
class Hero:
    def __init__(self,name,hp,damage):
        self.name = name
        self.hp = hp
        self.damage = damage
    def attack(self,other):
        other.hp -= self.damage
        print(f"{self.name} атаковал {other.name} → здоровье {other.name} = {other.hp}")
h1=Hero("Nurda",100,15)
h2=Hero("Nurzhan",100,10)
h1.attack(h2)
h2.attack(h1)
```

```python
#10
class ShopItem:
    def __init__(self,name,price,quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    def buy(self,n):
        if n>self.quantity:
            print("Not enough goods")
        else:
            self.quantity -= n
            print(f"Bought {n} goods")
    def restock(self,n):
        self.quantity += n
        print(f"Restock {n} goods ,Quantity={self.quantity}")
    def info(self):
        print(f"{self.name} Price: {self.price}, Quantity: {self.quantity}")

item=ShopItem("eggs",75,100)
item.buy(55)
item.buy(50)
item.restock(100)
item.info()
```

```python
class Spirit:
    def __init__(self,name,power):
        self.name = name
        self.power = power
        self.hp = 100
    def attack(self,target):
        pass
    def take_damage(self,amount):
        self.hp-=amount
        if self.hp<=0:
            self.hp=0
            print(f"{self.name} has faded")
class FireSpirit(Spirit):
    def attack(self,target):
        damage=self.power*2
        if isinstance(target,WaterSpirit):
            damage=damage/2
        else:
            damage=self.power
        target.hp-=damage
        print(f"{self.name}  attacks {target.name} and takes {damage} damage")
class WaterSpirit(Spirit):
    def attack(self,target):
        damage=self.power+3
        if isinstance(target,EarthSpirit):
            damage=damage*1.5
        else:
            damage=self.power
        target.hp-=damage
        print(f"{self.name} attacks {target.name} and takes  {damage} damage")
class EarthSpirit(Spirit):
    def attack(self,target):
        if isinstance(target,FireSpirit):
            damage=self.power*3
        else:
            damage=self.power
        target.hp-=damage
        print(f"{self.name}  attacks {target.name} and takes {damage} damage")


fire=FireSpirit("Fire",10)
water = WaterSpirit("Water",8)
earth=EarthSpirit("Earth",9)
fire.attack(water)
print(water.hp)
earth.attack(water)
print(water.hp)
water.attack(earth)
water.attack(fire)
print(fire.hp)
print(earth.hp)
```
