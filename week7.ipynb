```python
#dict7.
data={"apple":3,"kiwi":2,"banana":8}
pairs=[]
for key in data:
    pairs.append((key,data[key]))
for i in range(len(pairs)):
    for j in range(i+1,len(pairs)):
        len_i=len(pairs[i][0])
        len_j=len(pairs[j][0])
        val_i=pairs[i][1]
        val_j=pairs[j][1]
        if len_i>len_j or (len_i==len_j and val_i>val_j):
            pairs[i],pairs[j]=pairs[j],pairs[i]
print(pairs)
```

```python
#tuple6.
points=[(0,0),(0,3),(4,0),(1,2)]
best=()
maxarea=0
for i in range(len(points)):
    for j in range(i+1,len(points)):
        for k in range(j+1,len(points)):
            x1,y1=points[i]
            x2,y2=points[j]
            x3,y3=points[k]
            area=abs(x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
            if area>maxarea:
                maxarea=area
                best=(points[i],points[j],points[k])
print(best)
```
